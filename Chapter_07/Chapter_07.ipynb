# Classes and Interfaces

-   Python supports Object-Oriented Programming
    -   Including inheritance, polymorphism and encapsulation
-   Writing classes often involves defining their interfaces and
    relationships
-   Classes and inheritance let you express behaviours through objects
    -   Can improve and expand functionality over time
    -   Flexible in an environment of changing requirements
-   Python is also multi-paradigm
-   We can also write functional style code
    -   Functions are first class objects
    -   They can be treated like regular variables
-   Mixing paradigms is often more powerful than a single one
    -   See the common idea of the “imperative shell, functional core”
