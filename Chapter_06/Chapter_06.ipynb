# Chapter 6: Comprehensions and Generators

-   Many programs are built around processing collections
    -   e.g. lists, dictionary key, value pairs, sets
-   *Comprehensions* are a special syntax for for quickly iterating and
    creating collections
    -   They enhance readability
-   *Generators* enable a continuous stream of values to be returned via
    a function
    -   Generator results can be used anywhere a iterator is expected
    -   Generators improve readability, improve performance, reduce
        memory usage, and simplify implementations
